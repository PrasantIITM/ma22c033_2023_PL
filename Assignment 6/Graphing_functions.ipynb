```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
fig = plt.figure()
ax = plt.axes()
x=np.linspace(0,10,10000)
y=np.linspace(0,10,10000)
ax.plot(x,np.sin(x));
plt.title("the graph of sin(x)")
plt.xlabel("x");
plt.ylabel("sin(x)");
```

```python
import mpl_toolkits
from mpl_toolkits.mplot3d import Axes3D
```

```python
def f(x,y):
  return np.cos(np.sqrt(x**2+y**2))
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
x=np.linspace(-5,5,10000)
y=np.linspace(-5,5,10000)
X,Y=np.meshgrid(x,y)
Z=f(X,Y)
ax.plot_surface(X,Y,Z);
```
