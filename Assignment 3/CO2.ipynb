```python
import urllib
import matplotlib.pyplot as plt
import numpy as np
```

```python
from scipy import interpolate
```

```python
co2data='/content/Co2 data' #localfile
```

```python
ftp_url = 'ftp://aftp.cmdl.noaa.gov/products/trends/co2/co2_mm_mlo.txt'
```

```python
urllib.request.urlretrieve(ftp_url, co2data)
```

```python
relevant_data = np.genfromtxt(co2data, skip_header=72)
```

```python
averages = np.ma.masked_invalid(relevant_data[:, 4])
```

```python
plt.figure(figsize=(12, 7))
time=relevant_data[:,2]
# print(time)
plt.plot(time, averages,  linestyle='-', color='b')
plt.xlabel("Time")
plt.ylabel("CO2 Monthly Average")
plt.title("Monthly Averages of  CO2")
plt.grid(True)
plt.show()
```

```python
co2_rate_of_change = np.diff(averages)
# print(co2_rate_of_change)
time_rate_of_change = time[1:]
```

```python
plt.figure(figsize=(10, 6))
plt.plot(time_rate_of_change, co2_rate_of_change, linestyle='-', color='b')
plt.xlabel("Time")
plt.ylabel("Rate of Change of CO2")
plt.title(" Obeseravation of Rate of Change of CO2")
plt.grid(True)
plt.show()
```
