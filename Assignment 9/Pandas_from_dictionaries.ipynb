3. Convert d1, d2 (together) to a pandas object dd.

```python
d1 = {'a': 1, 'b': 2}
d2 = {'b': 3, 'c': 4}
```

```python
#Convert d1, d2 (together) to a pandas object dd.
dd=pd.DataFrame({'d1':d1,'d2':d2})
dd
```

5. Give examples of implicit and explicit indexing on dd.

```python
#Explicit Indexing: Explicit indexing refers to using the labels or names assigned to the rows and columns to access the data.
print(dd.loc['a', 'd1'])  # Accessing a specific element using label
```

```python
print(dd.loc['a':'b', 'd1'])  # Accessing a range of rows and a specific column using labels
```

```python
#Implicit Indexing:Implicit indexing refers to the default integer-based indexing of rows and columns in a DataFrame.
print(dd.iloc[0, 0])  # Accessing a specific element using integer-based indexing
```

```python
print(dd.iloc[0:2, 0])  # Accessing a range of rows and a specific column using integer-based indexing
```

6. Forward fll and backfll missing values using various axis.

```python
# Forward fill missing values along rows (axis=0)
dd_filled_rows = dd.ffill(axis=0)
print("Forward fill along rows (axis=0):")
print(dd_filled_rows)
```

```python
# Backward fill missing values along rows (axis=0)
dd_filled_rows_bfill = dd.bfill(axis=0)
print("\nBackward fill along rows (axis=0):")
print(dd_filled_rows_bfill)
```

```python
# Forward fill missing values along columns (axis=1)
dd_filled_columns = dd.ffill(axis=1)
print("\nForward fill along columns (axis=1):")
print(dd_filled_columns)
```

```python
# Backward fill missing values along columns (axis=1)
dd_filled_columns_bfill = dd.bfill(axis=1)
print("\nBackward fill along columns (axis=1):")
print(dd_filled_columns_bfill)
```
