```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn; seaborn.set();
from scipy.integrate import solve_ivp
import numpy as np
```

```python
def ppm_f(t,y,alpha,beta,sigma,gamma):
  r,x=y
  return np.array([alpha*r-beta*r*x,
                   sigma*r*x-gamma*x])
```

```python
alpha,beta,sigma,gamma=1.1,0.4,0.1,0.1
```

```python
sol = solve_ivp(ppm_f, [0, 70], [5,2],
                rtol=1e-6, args=(alpha,beta, sigma, gamma))
```

```python
fig = plt.figure(); ax = fig.gca()
curves = ax.plot(sol.t, sol.y.T)
ax.legend(curves, ['r', 'x']);
```

```python
def random_intial_condition():
  r_0=np.random.uniform()
  x_0=np.random.uniform()
  return [r_0,x_0]
fig = plt.figure(figsize=(8,8)); ax = fig.gca()
for i in range(10):
  intial_condition=random_intial_condition()
  sol = solve_ivp(ppm_f, [0, 70], intial_condition,
                rtol=1e-6, args=(alpha,beta, sigma, gamma),t=np.linspace(0,70,1000))

  curves = ax.plot(sol.t, sol.y.T)



ax.legend();
```

```python
def pp_plot(initial_cond=[5,2]):
  plt.figure(figsize=(8, 8))
  initial_conditions = initial_cond
  sol = solve_ivp(ppm_f, [0,70], initial_conditions, args=(alpha, beta, sigma, gamma), t_eval=np.linspace(0, 70, 1000))

  plt.plot(sol.y[1], sol.y[0], label=f'Initial: ({initial_conditions[1]:.2f}, {initial_conditions[0]:.2f})')

  plt.xlabel('x')
  plt.ylabel('r')
  plt.legend()
  plt.title('Phase Portrait')
  plt.show()

pp_plot()
```
